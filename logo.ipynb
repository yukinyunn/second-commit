```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Axesを作成する関数で　polar='True' を指定することで、極座標系の図を作成できる
ax = plt.subplot(polar='True')

n = 10

#プロット点のx座標n個を2pi間に等間隔で用意
#np.deg2radで度とラジアンの相互変換
theta = np.arange(np.deg2rad(0), np.deg2rad(360), 2*np.pi/n)

l = np.random.rand(n) #プロットする値(barの長さ)を乱数で作成[-]
width = np.pi / 4 * np.random.rand(n) #幅も乱数で定義

bars = plt.bar(theta, l, width=width, bottom=0.0)

#色付け&透過
for bar in bars:
    bar.set_facecolor(plt.cm.jet(np.random.rand())) #色を乱数で生成
    bar.set_alpha(0.6) #透明度60%
    
ax.set_xticklabels([]) #x軸(角度)非表示
ax.set_yticklabels([]) #y軸(角度)非表示
plt.savefig('logo.png')
plt.show()
```
